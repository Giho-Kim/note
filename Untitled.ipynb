```python
api = wandb.Api()
entity = api.default_entity
project = api.projects(entity)[2].name
runs = api.runs(f"{entity}/{project}")
```

```python
project
```

```python
# from $run.config $df.columns
configs = ['tilt_beta', 'z_mix_ratio', 'tilt_temperature_start', 'tilt_temperature_end', 'seed', 'tilt_init_geom_ratio', 'tilt_candidate_multiplier', 'tilt_ridge_alpha']
columns = ['eval/reach_top_right/episode_reward_iqm', 'eval/reach_top_left/episode_reward_iqm',
           'eval/reach_bottom_right/episode_reward_iqm','eval/reach_bottom_left/episode_reward_iqm', 
           'eval/task_reward_iqm']
```

```python
import pandas as pd

rows = []
log_interval = 20000

for i, run in enumerate(runs):
    print(i, run.state, run.name)
    if run.state == "running":
        continue
    
    df = run.history()
    ckpt = int(run.files()[0].name[:-7]) // log_interval

    row_dict = {}

    for config in configs:
        row_dict[config] = run.config[config]

    for column in columns:
        row_dict[column] = df[column].iloc[ckpt]
        # 또는 df.loc[ckpt, column] 가능, index가 0,1,2...이면

    rows.append(row_dict)

result_df = pd.DataFrame(rows)
```

# D-LEVER-MAZE 

```python
api = wandb.Api()
entity = api.default_entity
project = api.projects(entity)[5].name
runs = api.runs(f"{entity}/{project}")
project
```

```python
# from $run.config $df.columns
configs = ['domain_name', 'tilt', 'tilt_ridge_alpha', 'tilt_temperature_start', 'seed']
columns = ['eval/task_reward_iqm','eval/reach_top_left/episode_reward_iqm',
'eval/reach_top_right/episode_reward_iqm',
'eval/reach_bottom_left/episode_reward_iqm',
'eval/reach_bottom_right/episode_reward_iqm',
           ]
col = 'eval/task_reward_iqm'
```

```python
import pandas as pd

rows = []
log_interval = 20000

for i, run in enumerate(runs):
#     break
    df = run.history()
    
    if run.state == "running" or run.config["domain_name"] != 'point_mass_maze' or len(df) < 50:
        continue
    print(i, run.state, run.name, run.config["domain_name"])

    ckpt = df[col].argmax()
    

    row_dict = {}

    for config in configs:
        row_dict[config] = run.config[config]

    for column in columns:
        row_dict[column] = df[column].iloc[ckpt]

    rows.append(row_dict)

result_df = pd.DataFrame(rows)
```
